## 5.Seasonal Decomposing for checking trends and outliers

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

columns_to_decompose = ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School', 'Other']

df_recent = df_news.last('365D')

for col in columns_to_decompose:
    series = df_recent[col].dropna()
    
    result = seasonal_decompose(series, model='additive', period=7)
    
    fig, axes = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
    fig.suptitle(f'Seasonal Decomposition (Last 1 Year) of {col}', fontsize=14, fontweight='bold')
    
    result.observed.plot(ax=axes[0], color='tab:blue')
    axes[0].set_ylabel('Observed')
    axes[0].grid(True, linestyle='--', alpha=0.5)

    result.trend.plot(ax=axes[1], color='tab:orange')
    axes[1].set_ylabel('Trend')
    axes[1].grid(True, linestyle='--', alpha=0.5)

    result.seasonal.plot(ax=axes[2], color='tab:green')
    axes[2].set_ylabel('Seasonal')
    axes[2].grid(True, linestyle='--', alpha=0.5)

    result.resid.plot(ax=axes[3], color='tab:red')
    axes[3].set_ylabel('Residual')
    axes[3].set_xlabel('Date')
    axes[3].grid(True, linestyle='--', alpha=0.5)
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

## From the above graphs, we can see that there are outliers in the graph, we need to handle the outliers by using the rolling median


```python
for col in ['Peak Service', 'Other']:
    df_news[col] = df_news[col].mask(
        (df_news[col] > df_news[col].quantile(0.95)) |
        (df_news[col] < df_news[col].quantile(0.05)),
        df_news[col].rolling(window=3, center=True, min_periods=1).median()
    )
```

## 6.Cleaned Time series data

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(len(columns_to_decompose), 1, figsize=(12, 10), sharex=True)

for i, col in enumerate(columns_to_decompose):
    axes[i].plot(df_recent.index, df_recent[col], label=col, color='steelblue')
    axes[i].set_title(f'{col} - After Outlier Handling', fontsize=12, fontweight='bold')
    axes[i].set_ylabel('Passenger Count')
    axes[i].grid(True, linestyle='--', alpha=0.5)
    axes[i].legend(loc='upper left', fontsize=9)

axes[-1].set_xlabel('Date')

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.suptitle('Cleaned Time Series (After Outlier Handling)', fontsize=14, fontweight='bold')
plt.show()
```

# Forecasting the Columns using ARIMA

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
```

## ADF test for checking the dataset is stationary

```python
def adf_test(series, name=''):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f"\nADF Test for {name}")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.6f}")
    if result[1] <= 0.05:
        print("Data is stationary (no differencing needed).")
        return 0  # d = 0
    else:
        print("Data is non-stationary (differencing needed).")
        return 1  # d = 1
```

```python
columns = ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School']
d_values = {}

for col in columns:
    d = adf_test(df_news[col], col)
    d_values[col] = d
```

## ACF and PACF for getting the p and q value

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plot_acf(df_news['Local Route'], lags=30, ax=plt.gca())
plt.title('ACF - Local Route')

plt.subplot(1,2,2)
plot_pacf(df_news['Local Route'], lags=30, ax=plt.gca())
plt.title('PACF - Local Route')

plt.tight_layout()
plt.show()
```

## Forecasting with ARIMA model for the next 7 days

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pandas as pd

columns = ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School']

forecasts = {}

for col in columns:
    model = ARIMA(df_news[col], order=(1, 0, 1))
    results = model.fit()
    
    forecast = results.forecast(steps=7)
    forecast.index = pd.date_range(start=df_news.index[-1] + pd.Timedelta(days=1),
                                   periods=7, freq='D')
    forecasts[col] = forecast
    
    recent_data = df_news[col].iloc[-30:]
    
    plt.figure(figsize=(10,5))
    plt.plot(recent_data, label='Last 30 Days', color='steelblue')
    plt.plot(forecast, label='7-Day Forecast', linestyle='--', color='orange', marker='o')
    plt.title(f'{col} - Recent 30 Days & 7-Day Forecast')
    plt.xlabel('Date')
    plt.ylabel('Passenger Count')
    plt.xticks(rotation=45) 
    plt.legend()
    plt.tight_layout()
    plt.show()
    
for col, forecast in forecasts.items():
    print(f"\n Forecast for {col}:")
    print(forecast)
```
